Informació columnes:

Columnes d'informació:

- 1.CRIM: Taxa de criminalitat per càpita per municipis
- 2.ZN: de sòl residencial zonificat per a lots de més de 25.000 metres quadrats.
- 3.INDUS: hectàrees comercials no minoristes per ciutat

- 5.NOX: Concentració d'òxids nítrics(parts per 10 milions)
- 6.RM: nombre mitjà d'habitacions per habitatge
- 7.AGE: Proporció d'edat de les unitats de propietat construïdes abans de 1940
- 8.DIS: Distàncies ponderades fins a cinc centres d'ocupació de Boston
- 9.RAD: Índex d'accessibilitat a carreteres radials
- 10.TAX: Taxa d'impost sobre la propietat de valor total per 10.000 $
- 11.PTRATIO: Ràtio alumnes-professor per municipis
- 12.B: 1000(Bk - 0,63)^2 on Bk és la proporció de negres per ciutat
- 13.LSTAT: % d'estatus inferior de la població
- 14.MEDV: Valor mitjà de les cases ocupades pel propietari en 1.000 dòlars

## - Exercici 2
Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...).

Per poder estandaritzar les dades, és necessari que aquestes tinguin una distribució gaussina. El primer que farem serà veure si les dades numèriques del DataFrame tenen aquesta distribució. Fem la prova de Shapiro-Wilk a la primera columna.

```python
from scipy.stats import shapiro

stat, p = shapiro(X_train['CRIM'])
print('Estadistica de prova=%.3f, valor p=%.3f' % (stat, p))
if p > 0.05:
    print('Les dades sembla que tenen una distribució gaussiana')
else:
    print('Los dades sembla que no tenen una distribución gaussiana')
```

Aquesta columna, per començar doncs, no té una distribució gaussiana. Grafiquem les dades numèriques amb l'histograma:

```python
#seleccionem les columnes numèriques float (les altres són categòriques)
numeric_cols = X_train.select_dtypes(include=['float64'])

#Creem histogrames de totes les columnes numèriques
numeric_cols.hist(bins=10, figsize=(10,10))
plt.show()
```

Veiem que en general no tenen una distribució normal, però podem comprovar la columna de TAX, que s'hi pot apropar:

```python
stat, p = shapiro(X_train['TAX'])
print('Estadística de prueba=%.3f, valor p=%.3f' % (stat, p))
if p > 0.05:
    print('Les dades sembla que tenen una distribució gaussiana')
else:
    print('Los dades sembla que no tenen una distribución gaussiana')
```

Veiem que tampoc té una distribució gaussiana, és preferible doncs no estandaritzar el dataframe, perquè en Machine Learning podria donar resultats erronis d'aquesta manera.

A continuació hem dumificat la columna "RAD" (Índex d'accessibilitat a carreteres radials), ja que l'altre columna que s'hauria pogut dummificar, "CHAS", ja ho estava. D'aquesta manera es creen 9 columnes binàries de 0 i 1 de cada valor. D'aquesta manera, en Machine Learnig, no s'interpreten els valores dels registres categòrics com uns més grans que altres, per exemple i fa més fàcil l'aprenentage i l'anàlisi.

```python
X_train['RAD'].value_counts()
```

```python
df_housing = pd.get_dummies(X_train, columns = ["RAD"])
```

Creem polinomis. Serveixen per crear noves columnes que representin diferents potències de les columnes existents:

```python
from sklearn.preprocessing import PolynomialFeatures

# Seleccionem les columnes per transformar en polinomis
X = X_train[['CRIM', 'ZN']]

# Creem polinomis de grau 2
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Convertim a DataFrame i unim amb el DataFrame original
df_poly = pd.DataFrame(X_poly, columns=poly.get_feature_names_out(['CRIM', 'ZN']))
housing_poli = pd.concat([X_train, df_poly], axis=1)
```
