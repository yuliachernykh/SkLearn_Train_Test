Ara separem les columnes del target de la resta del DataFrame, que en aquest cas serà 'MEDV': Valor mitjà de les cases ocupades pel propietari en 1.000 dòlars.

```python
from sklearn.model_selection import train_test_split
features = ['CRIM', 'ZN', 'INDUS', 'CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
X = housing.loc[:, features]
y = housing.loc[:, ['MEDV']]    
```

Separem el data set dos, una part que serà del 80% per fer l'entrenament i el 20% restant per fer el test.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Com podem veure el total dels registres de l'entrenament correspon al 80% del DataFrame i el 20% del dataframe corresponent a les columnes menys el target.

## - Exercici 2
Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...).

Per poder estandaritzar les dades, és necessari que aquestes tinguin una distribució gaussina. El primer que farem serà veure si les dades numèriques del DataFrame tenen aquesta distribució. Fem la prova de Shapiro-Wilk a la primera columna.

```python
from scipy.stats import shapiro

stat, p = shapiro(X_train['CRIM'])
print('Estadistica de prova=%.3f, valor p=%.3f' % (stat, p))
if p > 0.05:
    print('Les dades sembla que tenen una distribució gaussiana')
else:
    print('Los dades sembla que no tenen una distribución gaussiana')
```

Aquesta columna, per començar doncs, no té una distribució gaussiana. Grafiquem les dades numèriques amb l'histograma:

```python
numeric_cols = X_train.select_dtypes(include=['float64'])

numeric_cols.hist(bins=10, figsize=(10,10))
plt.show()
```

Veiem que en general no tenen una distribució normal, però podem comprovar la columna de TAX, que s'hi pot apropar:

```python
stat, p = shapiro(X_train['TAX'])
print('Estadística de prueba=%.3f, valor p=%.3f' % (stat, p))
if p > 0.05:
    print('Les dades sembla que tenen una distribució gaussiana')
else:
    print('Los dades sembla que no tenen una distribución gaussiana')
```

Veiem que tampoc té una distribució gaussiana, és preferible doncs no estandaritzar el dataframe, perquè en Machine Learning podria donar resultats erronis d'aquesta manera.

A continuació, esbrinarem si les variables tenen outliers. Visualitzarem les variables amb gràfics de boxplot.

```python
import matplotlib.pyplot as plt

def boxplots_por_columna(df):
    for col in df.select_dtypes(include=[float]).columns:
        df[[col]].boxplot()
        plt.show()
```

Podem observar que les variables: CRIM, ZN, RM, DIS, PTRATIO, B, LSTAT tenen outliers i que INDUS, NOX, AGE i TAX no en tenen. A les que tenen outliers, els aplicarem el RobustScaler i a les altres, la Normalització.

```python
cols = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'TAX',
       'PTRATIO', 'B', 'LSTAT']

fig, axs = plt.subplots(nrows=len(cols), ncols=2, figsize=(12, 20))

for i, col in enumerate(cols):
    axs[i, 0].hist(X_train[col], bins=30, alpha=0.5, color='blue', label='Train')
    axs[i, 0].set_title(col)
    axs[i, 0].legend()
    
    axs[i, 1].hist(X_test[col], bins=30, alpha=0.5, color='orange', label='Test')
    axs[i, 1].set_title(col)
    axs[i, 1].legend()

plt.tight_layout()
```

A les gràfiques es poden apreciar els canvis d'escala.
